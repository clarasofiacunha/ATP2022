```python
#TPC3

import random

def menu():
    print('''Menu:
(1) Criar lista aleatória
(2) Criar lista
(3) Soma da lista
(4) Média da lista
(5) Maior elemento da lista
(6) Menor elemento da lista
(7) Verificar se a lista está Ordenada por ordem crescente
(8) Verificar se a lista está Ordenada por ordem decrescente
(9) Procurar um elemento na lista
(0) Sair''')
    return

def criarLista(a):
    lista1=[]
    while a>0:
        lista1.append(random.randint(1, 101))
        a= a-1
    return lista1

def leLista(a):
    lista2=[]
    i=0
    while a>0:
        num=int(input("Introduza o "+str(i+1)+" elemento:"))
        lista2.append(num)
        i=i+1
        a=a-1
    return lista2

def somaLista(lista):
    soma=0
    for num in lista:
        soma=soma+num
    return soma

def mediaLista(lista):
    media=somaLista(lista)/len(lista)
    return media

def maiorLista(lista):
    maior=lista[0]
    for i in lista:
        if i>maior:
            maior=i
    return maior

def menorLista(lista):
    menor=lista[0]
    for i in lista:
        if i<menor:
            menor=i
    return menor

def crescenteLista(lista):
    valor=True
    i=0
    while i+1<len(lista):
        if lista[i]>lista[i+1]:
            valor=False
        i=i+1
    return valor

def decrescenteLista(lista):
    valor=True
    i=0
    while i+1<len(lista):
        if lista[i]<lista[i+1]:
            valor=False
        i=i+1
    return valor

def pos(lista, elem):
    posLista=[]
    for i in range(len(lista)):
        if lista[i]==elem:
            posLista.append(i+1)
    if len(posLista)==0:
        posLista.append(-1)
    return posLista

menu()
s=int(input("Selecione uma opção"))
listaRes=[]

while s!=0:
    if s==1:
        listaRes.clear()
        listaRes=criarLista(int(input("Quantos elemntos quer na lista?")))
    if s==2:
        listaRes.clear()
        listaRes=leLista(int(input("Quantos elementos quer na lista?")))
    if s==3:
        print("A soma dos elementos da lista é:",somaLista(listaRes))
    if s==4:
        print("A média dos elementos da lista é:",mediaLista(listaRes))
    if s==5:
        print("O maior elemento da lista é:",maiorLista(listaRes))
    if s==6:
        print("O menor elemento da lista é:",menorLista(listaRes))
    if s==7:
        if crescenteLista(listaRes)==True:
            print("Sim. A lista está ordenada por ordem crescente.")
        else:
            print("Não. A lista nãp está ordenada por ordem crescente.")
    if s==8:
        if decrescenteLista(listaRes)==True:
            print("Sim.A lista está ordenada por ordem decrescente.")
        else:
            print("Não. A lista não está ordenada por ordem decrescente")
    if s==9:
        posLista=pos(listaRes,int(input("Qual é o elemento que procura?")))
        if posLista[0]==-1:
            print("O elemento que procura não se encontra na lista.")
        else:
            print("O elemento que procura encontra-se na(s)",posLista,"posição(ões).")
    s=int(input("Selecione uma opção"))
print("")
print("A sua lista é:", listaRes)
```
