```python
# Janela Consultar Indivíduo
def janelaConsulta(bd):
    sg.theme('LightGrey1')
    layout = [
    [sg.Text('Procurar Utilizador')],
    [sg.Text('Nome ou BI/CC:', size=(18,1)),sg.InputText(key='-procura',size=(20))],
    [sg.Button('Submeter')]
    ]
    
    window = sg.Window('Procurar Utilizador', layout)
    
    stop=False
    while not stop:
        event, values=window.read()
        if event == sg.WIN_CLOSED:
            stop = True
        else:
            procura = values['-procura']
            m.consulta(bd, procura)
            stop = True
    window.close()
```
